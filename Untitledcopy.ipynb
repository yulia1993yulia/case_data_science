```python
df2=df1.groupby(['order_id','warehouse_name', 'highway_cost'])['quantity'].sum().reset_index() #группируем по заказу, складу, стоимости доставки, суммируем по полю кол-во продуктов в опр заказе
df2
```

```python
df2['tarif']=df2['highway_cost']/df2['quantity'] #добавялем столбец с расчетом тарифа для каждого склада (повторяется)
df2
```

```python
df31=df2.drop_duplicates().set_index('warehouse_name') #удаляем дубликаты, чтобы получить таблицу склад, тариф
df31
```

```python
#пункт2
#Найти суммарное количество , суммарный доход , суммарный расход и суммарную прибыль для каждого товара (представить как таблицу со столбцами
#'product', 'quantity', 'income', 'expenses', 'profit')
```

```python
df4=df1.groupby(['product', 'warehouse_name', 'price'])['quantity'].sum().reset_index() #группируем таблицу df1 по полям продукт, склад, цена и подсчитываем суммарное кол-во каждого продукта, 
#заказанного на определенном складе
df4
```

```python
df4['income_product_warehouse']=df4['price']*df4['quantity'] #добавляем столбец прибыль с каждого продукта на каждом складе (кол-во*цена продукта)
df4
```

```python
df9=pd.merge(df4,df31['tarif'], on='warehouse_name', how='left') #подтягиваем для каждого склада тариф, чтобы подсчитать стоимость доставки каждого вида продукта на определенном складе
df9
```

```python
df9['expenses_product_warehouse']=df9['tarif']*df9['quantity'] #добавялем столбец расход для продукта на определенном складе (тариф*кол-во)
df9
```

```python
df10=df9.groupby(['product'])['quantity','income_product_warehouse','expenses_product_warehouse'].sum() #группируем по полю продукт, суммируем по полям кол-во, прибыль, расход
df10
```

```python
df10['profit']=df10['income_product_warehouse']-df10['expenses_product_warehouse'] #Находим прибыль каждого продукта
df10
```

```python
#пункт3
#Составить табличку со столбцами 'order_id' (id заказа) и 'order_profit' (прибыль полученная с заказа). А также вывести среднюю прибыль заказов
df1['order_product_price']=df1['price']*df1['quantity'] #добавили столбец суммарная стоимость каждого продукта в заказе = цена продукта*кол-во этого продукта в заказе
df1
```
